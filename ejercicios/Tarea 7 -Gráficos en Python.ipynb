```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 256

X = np.linspace(-np.pi, np.pi, n, endpoint=True)

Y = np.sin(2 * X)

plt.plot(X, Y + 1, color='blue', alpha=0.6, linewidth=1.3)
plt.fill_between(X,1, Y+1,color='mediumpurple', alpha=0.6)

plt.plot(X, Y - 1, color='blue', alpha=0.6,linewidth=1.3)
plt.fill_between(X, -1, Y-1, where= -1<=(Y-1), interpolate=True, color='mediumpurple',alpha=0.6)
plt.fill_between(X, -1, Y-1, where= -1>=(Y-1), interpolate=True, color='pink', alpha=0.8)

plt.yticks(())
plt.xticks(())
```

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

#T = np.arctan2(X,-Y)
T = np.arctan2(X,-Y)
plt.axes([0.25, 0.25, 0.95, 0.95])

plt.scatter(X,Y, s=56, c=T, cmap='jet', alpha=.5, edgecolors='black')

plt.xlim(-1.5, 1.5)
plt.xticks(())
plt.ylim(-1.5, 1.5)
plt.yticks(())
plt.show()
```

```python
n = 12

X = np.arange(n)

Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')

plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x, y in zip(X, Y1):
    plt.text(x, y+0.05, '%.2f' % y, ha='center', va='bottom')
plt.ylim(-1.25, +1.25)
for x, y in zip(X, -Y2):
    plt.text(x , y-0.05, '%.2f' % -y, ha='center', va='top')
```

# Ejercicio 4
A partir del siguiente código, intenta obtener el gráfico del dibujo llamado ejercicio 4 en el enunciado.

Pista: tendrás que modificar la Z.

Z = np.random.uniform(0, 1, 20)

plt.pie(Z)

```python
#Z = np.random.uniform(0, 1, 20)
n=20
Z= np.ones(20)
Z[-1] *= 2

plt.pie(Z,explode=Z*.05, colors = ['%f' % (i/float(n)) for i in range(n)], wedgeprops={'linewidth':0.5,"edgecolor":"black"})
plt.axis('equal')
plt.xticks(())
plt.yticks()

plt.show()
```
